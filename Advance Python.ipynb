```python
amount=int(input("Enter the total amount purchase:"))
if amount>=5000:
    coupon=input("Do you have a discount coupon?(yes/No):")
    if coupon.lower()=="yes":
        print("Discount applied! You saved 10%")
        discount=amount*10
        final_amount=amount-discount
        print("Final Amount to pay:",final_amount)
    else:
        print("No discount applied")
```
